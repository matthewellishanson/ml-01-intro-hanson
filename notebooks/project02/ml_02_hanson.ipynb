```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
disasters_df = pd.read_csv(
    '../../data/raw/global_extreme_weather_socioeconomic_2015_2025.csv'
)
disasters_df.columns
```

```python
disasters_df.isnull().sum()
```

```python
disasters_df = disasters_df.dropna()
print(f"Cleaned DataFrame shape: {disasters_df.shape}")
disasters_df.isnull().sum()
```

```python
disasters_df.describe()
```

```python
# target variable
y = disasters_df['total_affected']

# histogram of target variable
plt.figure(figsize=(10, 5))
sns.histplot(y, bins=30, kde=True)
plt.title("Histogram of Target Variable: total_affected")
plt.xlabel("total_affected")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
x_variables = ['duration_days', 'gdp_per_capita_usd', 'country_population']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for ax, col in zip(axes, x_variables, strict=True):
    sns.histplot(disasters_df[col], bins=30, kde=True, ax=ax)
    ax.set_title(f"Histogram of {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
